```python
def load_context():
    """Load context of all teh files in this directory"""
    context_dir = Path("/Users/dhanush/Desktop/Learning/langchain/Sales agent/context")
    context_dir.mkdir(exist_ok=True)

    all_content = ""

    for file_path in context_dir.glob("*"):
        if file_path.is_file():
            try:
                content = file_path.read_text(encoding = "utf-8")
                all_content += f"\n=== {file_path.name} ===\n{content}\n"
            except:
                pass
    return all_content.strip() or "No files found"

print(load_context())
print("loaded the context")



```
